## Identificando e removendo outliers

### Usando quartis para limite superior e inferior

```python
dados = pd.read_csv("/content/Aluguel_Residencial.csv",sep=";")
dados.head(3)
```

```python
dados.boxplot(["Valor"])
```

```python
dados[dados.Valor>500000]
```

```python
valor = dados["Valor"]
```

```python
q1 = valor.quantile(.25)
q3 = valor.quantile(.75)
iiq = q3 - q1
limite_inferior = q1 - (1.5*iiq)
limite_superior = q3 + (1.5*iiq)
```

```python
selecao = (valor >=limite_inferior) & (valor <= limite_superior)
```

```python
dados_limpo = dados[selecao]
```

```python
dados_limpo.boxplot(["Valor"])
```

### Remoção por grupo

```python
dados.boxplot(["Valor"], by = ["Tipo"])
```

```python
grupo_tipo = dados.groupby("Tipo")["Valor"]
```

```python
grupo_tipo.groups
```

```python
valor = grupo_tipo
```

```python
q1 = valor.quantile(.25)
q3 = valor.quantile(.75)
iiq = q3 - q1
limite_inferior = q1 - (1.5*iiq)
limite_superior = q3 + (1.5*iiq)
```

```python
q1,q3,iiq,limite_inferior,limite_superior
```

```python
limite_superior["Apartamento"]
```

```python
dados_limpos = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = dados["Tipo"] == tipo
  eh_dentro_limite = (dados["Valor"]>= limite_inferior[tipo]) & (dados["Valor"]<=limite_superior[tipo])
  selecao = eh_tipo & eh_dentro_limite
  dados_selecao = dados[selecao]
  dados_limpos = pd.concat([dados_limpos,dados_selecao])
```

```python
dados_limpos.boxplot(["Valor"], by = ["Tipo"])
```

```python
dados_limpos.to_csv("Aluguel_residencial_sem_outliers.csv",sep= ";",index = False)
```
